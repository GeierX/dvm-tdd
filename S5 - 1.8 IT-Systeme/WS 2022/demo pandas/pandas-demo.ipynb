```python
import pandas as pd

df = pd.read_csv("CovidFaelle_Timeline.csv", sep=";")
df['Time'] = pd.to_datetime(df['Time'], format="%d.%m.%Y %H:%M:%S")

#df['Time']

dfS = df.query("(Bundesland == 'Salzburg') and Time >= 20220101")
# fig = dfS.plot(x="Time", y="AnzahlFaelle").get_figure()
# fig.savefig("salzburg2022.png")

pS = dfS.plot(x="Time", y="AnzahlFaelle", label="Salzburg")
pW = df.query("Bundesland == 'Wien' and Time >= 20220101").plot(xlabel="", ylabel="Anzahl Fälle", x="Time", y="AnzahlFaelle", label="Wien", ax=pS)

pW.get_figure().savefig("combi.jpeg", dpi=100, bbox_inches='tight')
```
